```python
from transformers import pipeline
import torch
```

1. CLASIFICACIÓN DE TEXTO: ANÁLISIS DE SENTIMIENTOS

Modelo preentrenado basado en Transformers para la clasificación de texto según el sentimiento. Esto es útil para tareas como análisis de opiniones en redes sociales o reseñas de productos

```python
# Load the sentiment analysis pipeline
classifier = pipeline(
    'sentiment-analysis',
    model='distilbert-base-uncased-finetuned-sst-2-english'
)

# Example text in English
texts = [
    "I love studying natural language processing!",
    "Transformers are so confusing but amazing!",
    "This is the worst example I've ever seen."
]

# Analyze sentiments
results = classifier(texts)
for text, result in zip(texts, results):
    print(f"Text: {text}")
    print(f"Sentiment: {result['label']}, Confidence: {result['score']:.2f}")
```

2. RESPUERTA A PREGUNTAS

Usaremos un modelo de Transformers para responder preguntas basadas en un contexto dado

```python
# Ensure the model uses the GPU if available
qa_pipeline = pipeline(
    'question-answering',
    model="deepset/roberta-large-squad2",
    device=0  # Use GPU (if available)
)

# Context and questions
context = (
    "Hugging Face is a technology company that develops advanced tools "
    "for natural language processing (NLP). Among its products, the "
    "Transformers library stands out, widely used for tasks such as text classification, "
    "question answering, and text generation."
)

questions = [
    "What kind of tools does Hugging Face develop?",
    "What tasks can be performed with the Transformers library?"
]

# Answer questions
for question in questions:
    result = qa_pipeline(question=question, context=context)
    print(f"Question: {question}")
    print(f"Answer: {result['answer']} (confidence: {result['score']:.2f})\n")
```

3. GENERACIÓN DE TEXTO

Usaremos un modelo de transformers para generar texto basado en una entrada dada

```python
from transformers import pipeline

# Load the text generation pipeline
text_generator = pipeline('text-generation', model='gpt2')

# Initial prompt
prompt = "In the future, artificial intelligence will"

# Generate text
generated_text = text_generator(prompt, max_length=50, num_return_sequences=1)
print("Generated text:")
print(generated_text[0]['generated_text'])
```

LA TEMPERATURA

La temperatura ajusta la aleatoriedad de las predicciones:

- Valores bajos (p. ej., 0.2): Generan respuestas más conservadoras y deterministas.
- Valores altos (p. ej., 0.8): Generan respuestas más creativas y variadas.

```python
from transformers import pipeline

# Load the text generation pipeline
text_generator = pipeline('text-generation', model='gpt2')

# Initial prompt
prompt = "In the future, artificial intelligence will"

# Generate text with temperature control
generated_text = text_generator(
    prompt,
    max_length=200,
    num_return_sequences=3,
    temperature=0.9  # Adjust creativity level
)

print("Generated text:")
print(generated_text[0]['generated_text'])
```
